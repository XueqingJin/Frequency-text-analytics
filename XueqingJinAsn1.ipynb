```python
import nltk 
from nltk.corpus import stopwords
from nltk import FreqDist
import numpy as np
import pandas as pd
import nltk
from nltk import sent_tokenize, word_tokenize, pos_tag
```

```python
data = df=pd.read_csv('C:/Users/DELL/Desktop/fashion_data.csv')
text = df['review text'].tolist()
review = "".join(text)
```

# 1. Use a simple bag-of-words approach

```python
tokens = nltk.word_tokenize(review)
```

```python
words1 = [w.lower() for w in tokens]
words2 = [w for w in words1 if w.isalpha()]
```

```python
words4 = []
for w in tokens: 
    words4.append(w.lower())
```

```python
freq = FreqDist(words2)
sorted_freq = sorted(freq.items(),key = lambda k:k[1], reverse = True)
sorted_freq
```

```python
freq.plot(30)
```

# 2.Use a bag-of-words approach with stemming and stop words removal

```python
del stopwords
```

```python
#stopword removal
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
words_nostopwords = [w for w in words2 if w not in stopwords]
freq_nostw = FreqDist(words_nostopwords)
sorted_freq_nostw = sorted(freq_nostw.items(),key = lambda k:k[1], reverse = True)
sorted_freq_nostw
```

```python
#freq_nostw.plot(30)
```

```python
#Stemming  Use Porter Stemmer 
porter = nltk.PorterStemmer()
stem1 = [porter.stem(w) for w in words_nostopwords]
freq1 = FreqDist(stem1)
sorted_freq1 = sorted(freq1.items(),key = lambda k: k[1], reverse = True)
sorted_freq1
```

```python
freq1.plot(30)
```

# 3. Use POS approach and focus on all the noun forms (NN, NNP, NNS, NNPS)

```python
POS_tags = nltk.pos_tag(tokens) 
POS_tags
```

```python
POS_tag_listN = [(word,tag) for (word,tag) in POS_tags if tag.startswith('N')]
```

```python
tag_freq = nltk.FreqDist(POS_tag_listN)
sorted_tag_freq = sorted(tag_freq.items(), key = lambda k:k[1], reverse = True)
sorted_tag_freq
```

```python
tag_freq.plot(30)
```

# 4. Use POS approach and only focus on NNP

```python
POS_tags = nltk.pos_tag(tokens) 
POS_tags
```

```python
POS_tag_NNP = [(word,tag) for (word,tag) in POS_tags if tag =='NNP']
```

```python
tag_freq_NNP = nltk.FreqDist(POS_tag_NNP)
sorted_tag_freq_NNP = sorted(tag_freq_NNP.items(), key = lambda k:k[1], reverse = True)
sorted_tag_freq_NNP
```

```python
tag_freq_NNP.plot(30)
```
